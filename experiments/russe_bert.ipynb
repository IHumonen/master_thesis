## BERT

```python
def get_embedding(tokens, location):
    model.eval()
    with torch.no_grad():
        ids = tokenizer.convert_tokens_to_ids(tokens)
        ids = torch.tensor(ids).reshape((1, len(tokens)))
        embeds = model(ids)['last_hidden_state'].squeeze()
        word_embeds = embeds[location, :]
        
        return word_embeds.sum(dim=0).numpy()
```

```python
tokenizer = AutoTokenizer.from_pretrained('DeepPavlov/rubert-base-cased')
model = AutoModel.from_pretrained('DeepPavlov/rubert-base-cased')
```

```python
def get_word_location(target, tokens):
    current = ''
    current_indices = []
    for i, token in enumerate(tokens):
        if token[:2] == '##':
            current += token[2:]
            current_indices.append(i)
        else:
            current = token
            current_indices = [i]
        if i < len(tokens)-1:
            next_ = tokens[i+1]
            if next_[:2] != '##':
                if current == target:
                    return current_indices
        else:
            if current == target:
                return current_indices
    print(target, tokens)
    return 'not found'
```
